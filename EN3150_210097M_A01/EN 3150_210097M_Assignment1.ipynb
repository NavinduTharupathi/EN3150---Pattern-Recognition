```python
import numpy as np
import matplotlib . pyplot as plt
from sklearn . model_selection import train_test_split  #used to split datasets into random train and test subsets.
from sklearn . linear_model import LinearRegression
import seaborn as sns
import pandas as pd

# Generate 100 samples
n_samples = 100
# Generate X values ( uniformly distributed between 0 and 10)
X = 10 * np . random . rand ( n_samples , 1) #Generates n_samples (100) random numbers uniformly distributed between 0 and 1.:np.random.rand(n_samples, 1)
# Generate epsilon values ( normally distributed with mean 0 and standard deviation 15)
epsilon = np . random . normal (0 , 15 , n_samples ) #Generates n_samples (100) random values from a normal (Gaussian) distribution with a mean of 0 and a standard deviation of 15. This will act as the noise in the data.
# Generate Y values using the model Y = 3 + 3X + epsilon
Y = 3 + 2 * X + epsilon [: , np . newaxis ]
#epsilon[:, np.newaxis] adds the noise to each Y value. epsilon[:, np.newaxis] reshapes the epsilon array to ensure it's added correctly to the X array
```

```python
#np.random.randint(104): Generates a random integer between 0 and 103. This value is used as the random_state in the train_test_split function to ensure reproducibility
r = np . random . randint (104)
# Split the data into training and test sets (80% train , 20% test )
X_train , X_test , Y_train , Y_test = train_test_split (X , Y ,test_size =0.2 , random_state = r )
# Plot the data points
plt . figure ( figsize =(10 , 6) )
plt . scatter ( X_train , Y_train , alpha =1 , marker ='o', color ='red', label ='Training Data ')
plt . scatter ( X_test , Y_test , alpha =1 , marker ='s', color ='blue', label ='Testing Data ')
plt . show ()
```

2.

* r is a random integer generated between 0 to 103

* in train_test_split function training and testing data sets has been divided into two parts

* traing set has 80% of the data set and test set has 20% from the data set

* by specifying the random state in will keep the given data set seperation as a constant

* since we have given 104 random states this can produce 104 different training and test sets


```python
for i in range (10) : # Plotting 10 different instances
     X_train , X_test , Y_train , Y_test = train_test_split (X ,Y , test_size =0.2 , random_state = np . random . randint(104) )
     model = LinearRegression ()
     model . fit ( X_train , Y_train )
     Y_pred_train = model . predict ( X_train )
     plt . plot ( X_train , Y_pred_train , label = f'LR {i+1} ')
plt . xlabel ('X')
plt . ylabel ('Y')
plt . legend ()
plt . show ()
```

```python
num_dependent_variables = 1 if y.ndim == 1 else y.shape[1]
print("Number of Dependent Variables:", num_dependent_variables)
```

3.   We can't directly apply linear regression to this model.

*    Neet to handle the missing values in the data set.
*    Linear regression models require numerical values as the input, so if there is any categorical values we need to encode them into numerical values.
*    Need to apply feature scaling methods to standardize and normalize data to improve the model performance.

4. This approach might lead to a missalignment between X and Y indeces.

*    we need to combine x and y values into a single data frame and remove the NaN values. When using this method there won't be any missalignment in the data set.



```python
data = pd.concat([X, y], axis=1)
data.dropna(inplace=True)
```

```python
if 'Gender' in data.columns:
    data['Gender'] = data['Gender'].replace({'Male': 1, 'Female': 0})
else:
    print("The gender column was not found. Please check the column names.")
```

```python
if 'Age' in data.columns:
    data['Age'] = data['Age'].map({

    '18-20': 1,
    '21-25': 1,
    '26-30': 1,
    '21-30': 1,
    '31-40': 2,
    '41-50': 2,
    '51-60': 3,
    '>60': 3

    })
```

```python
if 'Ethnicity' in data.columns:
    data['Ethnicity'] = data['Ethnicity'].map({

        'White': 1,
        'Black or African-American': 2,
        'Asian': 3,
        'Multiracial': 4,
        'Hispanic/Latino': 5,
        'American Indian or Alaskan Native': 6,

    })
```

```python
X_cleaned = data[X.columns]
y_cleaned = data['aveOralM']

independent_features = ['Age', 'Humidity', 'Max1L13_1', 'T_LC1', 'RCC1']
X_cleaned = data[independent_features]
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y_cleaned, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both the training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

# Initialize and train the model
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
# Get the estimated coefficients
coefficients = model.coef_

# Get the intercept (optional)
intercept = model.intercept_

# Combine feature names with their corresponding coefficients
coeff_dict = dict(zip(X_cleaned.columns, coefficients))

# Display coefficients
print("Estimated Coefficients for the Independent Variables:")
for feature, coeff in coeff_dict.items():
    print(f"{feature}: {coeff:.4f}")

print(f"Intercept: {intercept:.4f}")
```

```python
X_cleaned = data[X.columns]
y_cleaned = data['aveOralM']

independent_features = ['Age', 'T_OR1', 'T_OR_Max1', 'T_FHC_Max1', 'T_FH_Max1']
X_cleaned = data[independent_features]

from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_cleaned, y_cleaned, test_size=0.2, random_state=42)

from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both the training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

from sklearn.linear_model import LinearRegression

# Initialize and train the model
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Get the estimated coefficients
coefficients = model.coef_

# Get the intercept (optional)
intercept = model.intercept_

# Combine feature names with their corresponding coefficients
coeff_dict = dict(zip(X_cleaned.columns, coefficients))

# Display coefficients
print("Estimated Coefficients for the Independent Variables:")
for feature, coeff in coeff_dict.items():
    print(f"{feature}: {coeff:.4f}")

print(f"Intercept: {intercept:.4f}")
```

```python
# Predict on the test set
y_pred = model.predict(X_test_scaled)

# Calculate RSS
RSS = np.sum((y_test - y_pred) ** 2)
print(f"Residual Sum of Squares (RSS): {RSS:.4f}")

# Number of observations and features
N = X_test.shape[0]
d = X_test.shape[1]

# Calculate RSE
RSE = np.sqrt(RSS / (N - d - 1))
print(f"Residual Standard Error (RSE): {RSE:.4f}")

# Calculate MSE
MSE = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {MSE:.4f}")

# Calculate R-squared
R2 = r2_score(y_test, y_pred)
print(f"R-squared (R²): {R2:.4f}")

# Add an intercept column to X_train_scaled for statsmodels
X_train_scaled_with_intercept = sm.add_constant(X_train_scaled)

# Fit the model using statsmodels
model_sm = sm.OLS(y_train, X_train_scaled_with_intercept).fit()

# Extract standard errors, t-statistics, and p-values
standard_errors = model_sm.bse
t_statistics = model_sm.tvalues
p_values = model_sm.pvalues

# Get the feature names along with 'const' for the intercept
features_with_intercept = ['Intercept'] + independent_features

# Print Standard Errors with feature names
print("Standard Errors for each feature:")
for feature, se in zip(features_with_intercept, standard_errors):
    print(f"{feature}: {se:.4f}")

# Print t-statistics with feature names
print("\nt-statistics for each feature:")
for feature, t in zip(features_with_intercept, t_statistics):
    print(f"{feature}: {t:.4f}")

# Print p-values with feature names
print("\np-values for each feature:")
for feature, p in zip(features_with_intercept, p_values):
    print(f"{feature}: {p:.4f}")
```
